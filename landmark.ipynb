```python
### jaffe 
#img, label, points, regions
data_jaffe = [[0 for _ in range(213)],[0 for _ in range(213)],[0 for _ in range(213)],[0 for _ in range(213)]] 
rootdir = './jaffe/'
classes = os.listdir(rootdir)
count = 0
index = [0,1,8,9,16,17,26,27,46,47]
for i in range(len(classes)):
    class_path = os.path.join(rootdir, classes[i])
    files = os.listdir(class_path) 
    print(classes[i])
    for j in range(1, len(files)//2+1): 
        # print(j)
        imgpath = os.path.join(class_path, classes[i].lower()+str(j)+'.jpg')
        # print(imgpath)
        img = cv2.imread(imgpath)
        data_jaffe[0][count] = img
        data_jaffe[1][count] = classes[i].lower()
        
        lmkpath = os.path.join(class_path, classes[i].lower()+str(j)+'.jpg.detect_ret')
        # print(lmkpath)
        ff = open(lmkpath, 'rb')
        obj = pickle.load(ff)
        ff.close()
        points = []
        regions = []
        for k in index:
            pointx = int(obj['points'][k]['x'])
            pointy = int(obj['points'][k]['y'])
            points.append((pointx, pointy))
            regions.append(img[pointy-24:pointy+24, pointx-24:pointx+24, :])            
        data_jaffe[2][count] = points
        data_jaffe[3][count] = regions
        count += 1

print(count)
# data_jaffe = np.array(data_jaffe)
# with open('./jaff.pkl', 'wb') as f:
#      pickle.dump(data_jaffe, f)

```

```python
import matplotlib.pyplot as plt
import itertools
import numpy as np
def plot_confusion_matrix(cm, classes,title='Confusion matrix for RAF-DB based on FFN(AlexNet)',cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
　　 cm:混淆矩阵值
　　 classes:分类标签
　　 """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],horizontalalignment="center",color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    
## alexnet  
cm = np.array([[265,6,5,20,11,2,20],[18,29,1,9,6,6,5],[5,2,79,18,18,9,29],
               [15,4,9,1044,26,9,78],[6,1,9,66,313,7,76],[12,2,5,25,6,96,16],
               [22,3,15,68,56,4,512]])
## vgg
# cm = np.array([[266,3,14,14,15,1,16],[14,36,2,9,6,12,4],[4,1,95,12,14,11,23],
#                [15,0,51,1049,17,8,45],[6,1,70,25,310,14,52],[8,2,15,9,1,118,9],
#                [21,0,89,42,28,4,496]])

s = np.sum(cm,axis=1).reshape(7,-1)
cm = cm/s*100
cm = np.around(cm, decimals=2)
classes = ['Surprise','Fear','Disgust','Happy','Sad','Angry','Neutral']
plot_confusion_matrix(cm, classes)
```
